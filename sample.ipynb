```python
# Import libraries
import json
import pandas as pd
import os
```

#### **1. Lấy data về user infor và movie infor**

#### ***User infor***

```python
# User Infor:
users_inf = {
    "user_id": [], 
    "user_name": [],
    "member_since": []
}
```

```python
# Read data in file.json
directory = "data/users_inf" #Lưu ý lấy path cho chuẩn
for filename in os.listdir(directory):
    if filename.endswith('.json'):
        filepath = os.path.join(directory, filename)
        # Open and load the JSON file
        with open(filepath, 'r') as file:
            data_json = json.load(file)
        
        for key in users_inf.keys():
            if key in data_json:
                users_inf[key].append(data_json[key])


users_inf_df = pd.DataFrame(users_inf)
users_inf_df
```

#### ***Movie Infor***

```python
# Movie Infor:
movies_inf = {
    "movie_id": [],
    "title": [], 
    "introduction": [], 
    "runtime": [], 
    "runtimeSeconds": [], 
    "rating": [], 
    "award": [], 
    "genre": [], 
    "releaseDate": [], 
    "releaseLocation": [], 
    "actors": [], 
    "directors": []
}
```

```python
# Read data in file.json
directory = "data/movies_inf" #Lưu ý lấy path cho chuẩn
for filename in os.listdir(directory):
    if filename.endswith('.json'):
        filepath = os.path.join(directory, filename)
        # Open and load the JSON file
        with open(filepath, 'r') as file:
            data_json = json.load(file)
        
        for key in movies_inf.keys():
            if key in data_json:
                movies_inf[key].append(data_json[key])


movies_inf_df = pd.DataFrame(movies_inf)
movies_inf_df
```

#### **2. Lấy data về user rating cho movie**

#### ***1 Movie - Nhiều Users***

```python
data_mur = {
    'user_id': [],
    'movie_id': [],
    'user_rating': []
}
```

```python
# Read data in file.json
directory = 'data/data_mur/'
for filename in os.listdir(directory):
    if filename.endswith('.json'):
        filepath = os.path.join(directory, filename)
        # Open and load the JSON file
        with open(filepath, 'r') as file:
            data_json = json.load(file)
        
        for key in data_mur.keys():
            if key in data_json:
                data_mur[key] = data_mur[key] + data_json[key]


mur_df = pd.DataFrame(data_mur)
mur_df
```

```python
mur_df['movie_id'].value_counts()
```

#### ***1 User - Nhiều Movies***

```python
data_umr = {
    'user_id': [],
    'movie_id': [],
    'user_rating': []
}
```

```python
# Read data in file.json
directory = 'data/data_umr/'
for filename in os.listdir(directory):
    if filename.endswith('.json'):
        filepath = os.path.join(directory, filename)
        # Open and load the JSON file
        with open(filepath, 'r') as file:
            data_json = json.load(file)
        
        for key in data_umr.keys():
            if key in data_json:
                data_umr[key] = data_umr[key] + data_json[key]


umr_df = pd.DataFrame(data_umr)
umr_df
```

```python
umr_df['user_id'].value_counts()
```

#### **Note**:
Trong quá trình crawl có nhiều trang movie không có rating, hay là user không có thông tin,... nên có nhiều dữ liệu bị mất, nên mọi người lưu ý trong quá trình làm

#### ***Gộp lại tất cả và loại bỏ duplicated***

```python
data_df = pd.concat([mur_df, umr_df])
data_df = data_df.reset_index(drop = True)
data_df
```

```python
data_df = data_df.drop_duplicates()
data_df
```
