```python
# Test Regex Pattern
# List of strings
strings = [
    '/title/tt0297626/?ref_=ur_urv',
    '/title/tt0297626/'
]

# Regex pattern to match 'ur' followed by digits
pattern = r'/title/(\w+)/?'

for string in strings:
    # Find all matches of the pattern in the string
    matches = re.findall(pattern, string)
    
    # Print the matches
    if matches:
        print(matches[0])
```

```python
# user_name = []
# user_id = []
user_name = []
user_id = set()
for content in container:
    name_tag = content.find('span', class_ = "display-name-link")
    # user_name.append(name_tag.text)
    # user_id.append(name_tag.find('a')['href'][6:-1])
    if name_tag != None:
        user_name.append(name_tag.text)
        user_id_re = re.findall(pattern, str(name_tag.find('a')['href']))
        user_id.add(user_id_re[0])
user_name
```
