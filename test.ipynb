##### Test: Lưu file json chứa movie_id và user_id

```python
data = {}
```

```python
page = requests.get("https://www.imdb.com/title/tt0111161/reviews/?ref_=tt_ql_2")
soup = BeautifulSoup(page.text)
soup
```

```python
container = soup.find_all('div', class_ = "lister-item-content")
container
```

```python
# Test Regex Pattern
# List of strings
strings = [
    '/user/ur004337/?ref_=tt_ur',
    '/user/ur004337/'
]

# Regex pattern to match 'ur' followed by digits
pattern = r'\/user\/(ur\d+)'

for string in strings:
    # Find all matches of the pattern in the string
    matches = re.findall(pattern, string)
    
    # Print the matches
    if matches:
        print(matches[0])
```

```python
name_tags = soup.find_all('span', class_ = "display-name-link")
# user_name = []
# user_id = []
user_name = set()
user_id = set()
for name_tag in name_tags:
    # user_name.append(name_tag.text)
    # user_id.append(name_tag.find('a')['href'][6:-1])
    user_name.add(name_tag.text)
    user_id_re = re.findall(pattern, str(name_tag.find('a')['href']))
    user_id.add(user_id_re[0])

user_id
```

```python
data['user_id'] = list(user_id)
```

```python
data
```

```python
# Save data into file json
file_path = "data_test.json"
with open(file_path, 'w') as json_file:
    json.dump(data, json_file)
```
